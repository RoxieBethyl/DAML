```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline, BSpline
```

```python
def get_hist_info(data, bins=30):
    """
    Derive the appropriate numbers in the data

    Parameters:
    -----------
    data: ndarray
        Array of the x coordinates of the data
    bins (optional): Integer
        Bins of the histogram; changes the number of scatter points to outputs

    Returns:
    --------
    bins: 
        X values of the dataset in the histogram (taken as the of each bin)
    n: 
        Y values of the dataset (frequency or occurences in each bin)
    """
    # Here shifting x values to the center of the bins and
    # getting the height of the bin for y values
    
    n, bins = np.histogram(data, bins=bins)
    bins = [(bins[i] + bins[i+1])/2. for i in range(len(bins)-1)]
    return np.array(bins), np.array(n)
```

```python
path = os.getcwd() + '/GeantExample1Part2'
bins = 30
```

```python
with open(path + '/output_nt_Behind_Shield_Original.csv') as f:
    b_shield = f.readlines()
    b_shield = [line for line in b_shield if not line[0].startswith('#')]
    b_shield = [line.split('\n')[0] for line in b_shield]
    b_shield = np.array(b_shield, dtype=float)

b_shield
```

```python
with open(path + '/output_nt_Shield_Original.csv') as f:
    shield = f.readlines()
    shield = [line for line in shield if not line[0].startswith('#')]
    shield = [line.split('\n')[0] for line in shield]
    shield = np.array(shield, dtype=float)

shield
```

```python
plt.hist(shield, bins=bins, label='Lead Shield')
plt.hist(b_shield, bins=bins, label='Aluminium Object Behind Shield')
plt.xlabel('Energy Deposited (MeV)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Plottting the same graph with smooth curve
x_shield_plt, y_shield_plt = get_hist_info(shield, bins=bins)
x_b_shield_plt, y_b_shield_plt = get_hist_info(b_shield, bins=bins)

x_shield_plt_smooth = np.linspace(x_shield_plt.min(), x_shield_plt.max(), 300)
y_shield_plt_smooth = make_interp_spline(x_shield_plt, y_shield_plt)(x_shield_plt_smooth)

x_b_shield_plt_smooth = np.linspace(x_b_shield_plt.min(), x_b_shield_plt.max(), 300)
y_b_shield_plt_smooth = make_interp_spline(x_b_shield_plt, y_b_shield_plt)(x_b_shield_plt_smooth)

plt.plot(x_shield_plt_smooth, y_shield_plt_smooth, label='Shield')
plt.plot(x_b_shield_plt_smooth, y_b_shield_plt_smooth, label='Object Behind Shield')

plt.legend()
plt.show()
```

```python
orginal_ratio = np.average(b_shield/shield)
print("The average ratio of the deposited energies of the two objects is:", orginal_ratio)
```

The sheild has more energy deposited on the it than the object behind it as the ratio is less than 1 when calculating it as the energy on the object over the energy on the sheild.
