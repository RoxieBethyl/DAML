```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def expgenerator(mean, size, interval):
    data = []
    while len(data) <= size:
        randNo = np.random.exponential(mean)
        if interval[0] <= randNo <= interval[1]:
            data.append(randNo)
            
    return data
```

```python
tau = 2.2 #mircoseconds
#s = np.random.normal(mean, sigma, 1000)
decayt = expgenerator(tau, 1000, (0, 10))
decayt
```

```python
plt.hist(decayt)
```

```python
np.mean(decayt)
```

```python
results = []
NosExps = 500 #Number of Experiments
for i, exps in enumerate(range(0, NosExps)):
    dataset = expgenerator(tau, 1000, (0, 10))
    results.append({
        'data': dataset,
        'mean': np.mean(dataset)
    })
```

```python
meanL = []
for i in results:
    meanL.append(i['mean'])

plt.hist(meanL, bins=20)
plt.axvline(tau, c='r', label="Mean Lifetime")
plt.axvline(np.mean(meanL), c='orange', label="Estimated Lifetime")
plt.legend()
plt.show()
```

```python
bias = tau - np.mean(meanL)
precision = np.std(meanL)
prec_bais = precision/np.sqrt(NosExps) #Precision on the Bias
```

```python
print("The bias is" , bias)
print("The precision is" ,precision)
print("The precison on the bias is" ,prec_bais)
```

The bias tells us how far the experimental mean has strayed from the mean lifetime of the experiment.
The precision is the standard deviation of the experimental simulation and for this exercise, it was done by taking the standard devation of the means found for each of the 500 experiments.
The precision on the bias tells us the error on the experimental and is found as the precision divided by the sqaure root of the number of experiments.
